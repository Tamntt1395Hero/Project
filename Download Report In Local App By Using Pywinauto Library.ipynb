```python
def get_report(to_date: str):
    for File in Files:
        for Season in PLM_Season:
            Excel = LoadingPath+"\\"+File
            app = Application(backend="uia").start(r'{} "{}"'.format(program_path,Excel)).connect(path = program_path)
            time.sleep(5)
            try: 
                dlg = Desktop(backend="uia").window(title_re=".*Excel.*")
                dlg.maximize()
                time.sleep(3)
            except TimeoutError or ValueError:
                print(TimeoutError)
                print(ValueError)
                os.system('pause')
            print("Open application successfully")
            print(f"{datetime.datetime.now()}-[get_report] login")
            # try:
            dlg.child_window(title="Add-ins", auto_id="TabAddIns", control_type="TabItem").click_input()
            dlg.child_window(title="BEx Analyzer", control_type="MenuItem").click_input()
            dlg.child_window(title="Change Variable Values", control_type="MenuItem").click_input()
            dlg.child_window(title="BW - Production", control_type="ListItem").click_input(button='left', double=True)
            time.sleep(1)
            dlg.child_window(title="User", auto_id="1000", control_type="Edit").set_text(USERNAME)
            time.sleep(1)
            dlg.child_window(title="Password", auto_id="1002", control_type="Edit").set_text(PASSWORD)
            time.sleep(1)
            dlg.child_window(title="OK", auto_id="1", control_type="Button").click()
            print(f"{datetime.datetime.now()}-[get_report] login sucessfully")
            dlg.child_window(title="PLM key date:(*)", auto_id="1001", control_type="Edit").wait("exists enabled visible ready", timeout= 60)
            # try: 
            dlg.child_window(title="PLM key date:(*)", auto_id="1001", control_type="Edit").set_text("")
            time.sleep(1)
            dlg.child_window(title="PLM key date:(*)", auto_id="1001", control_type="Edit").type_keys(Yesterday)
            dlg.child_window(title="Marketing Division:(*)", auto_id="1001", control_type="Edit").set_text("")
            dlg.child_window(title="Marketing Division:(*)", auto_id="1001", control_type="Edit").type_keys(Marketing_Division)
            time.sleep(1)
            dlg.child_window(title="PLM Season:(*)", auto_id="1001", control_type="Edit").set_text("")
            dlg.child_window(title="PLM Season:(*)", auto_id="1001", control_type="Edit").type_keys(Season)
            time.sleep(1)
            dlg.child_window(title="OK", auto_id="btnOk", control_type="Button").click()
            # except ValueError:
            #     print(ValueError)
            #     os.system('pause')
            dlg.child_window(title="File Tab", auto_id="FileTabButton", control_type="Button").wait("exists enabled visible ready active", timeout= 600)

            try: 
                dlg.child_window(title="File Tab", auto_id="FileTabButton", control_type="Button").click()
                time.sleep(1)
                # Use keyboard shortcut to fast access to save as file in folder
                #%FA means use ALT for % to show File-->Save a Copy then use type_keys(O) to open folder
                dlg.type_keys('%FA')
                time.sleep(1)
                dlg.type_keys('O')
            except ValueError:
                print(ValueError)
                os.system('pause')
                time.sleep(2)                

            if not os.path.exists(SavingPath+"\\"+Today):
                os.makedirs(SavingPath+"\\"+Today)
            try: 
                dlg.child_window(title="File name:", auto_id="1001", control_type="Edit").set_text("")
                time.sleep(1)
                dlg.child_window(title="File name:", auto_id="1001", control_type="Edit").type_keys(SavingPath+"\\"+Today+"\\"+File.split(".xlsm")[0]+Today+".xlsm", with_spaces = True)
                time.sleep(1)
                dlg.child_window(title="Save", auto_id="1", control_type="Button").click()
                print(f"{datetime.datetime.now()}-[get_report] File Save - {File}")
            except ValueError:
                print(ValueError)
                os.system('pause')
            time.sleep(10)
            try: 
                dlg.type_keys('%F')
                time.sleep(1)
                dlg.type_keys('C')
            except ValueError:
                print(ValueError)
                os.system('pause')
            time.sleep(10)

            app.kill()
            time.sleep(5)
        #     dlg.print_control_identifiers()
```

```python
if __name__ == "__main__":
    get_report(to_date=datetime.datetime.now().strftime("%d.%m.%Y %H%M"))
```
